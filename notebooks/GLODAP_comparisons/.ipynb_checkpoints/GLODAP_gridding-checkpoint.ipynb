```python
lat, lon = ds["G2latitude"].to_xarray(), ds["G2longitude"].to_xarray()
years = ds["G2year"].to_xarray()
months = ds["G2month"].to_xarray()
days = ds["G2day"].to_xarray()
minutes = ds["G2minute"].to_xarray()
times = years + (months / 12) + (days / 365) + (minutes / 525600)
cfc12 = ds["G2cfc12"].to_xarray()
depths = ds["G2depth"].to_xarray()
cast_num = ds["G2cast"].to_xarray().astype("int").astype("str")
cruise_num = ds["G2cruise"].to_xarray().astype("int").astype("str")
station_num = ds["G2station"].to_xarray().astype("int").astype("str")

cfc12.values[cfc12.values == -9999] = np.nan
cast_codes = ds["G2expocode"].to_xarray().astype("str")


tmp = np.char.add(cast_codes.values, "_")
tmp = np.char.add(tmp, cruise_num)
tmp = np.char.add(tmp, "_")
tmp = np.char.add(tmp, station_num)
tmp = np.char.add(tmp, "_")
tmp = np.char.add(tmp, cast_num)

all_casts = copy.deepcopy(cast_codes)
all_casts.values = tmp

unique_years = np.unique(years)

LONS_reg = np.arange(-180, 180, 2)
LATS_reg = np.arange(-90, 90, 2)

gridded_z = np.arange(100, 6000, 75)
```

```python
cfcs_gridded_nearest = np.nan * np.zeros((len(LONS_reg), len(LATS_reg), len(gridded_z), len(unique_years)))
```

```python
cfc_gridded = xr.DataArray(cfcs_gridded_nearest, dims=("lon", "lat", "depth", "time"), 
             coords={"lon": LONS_reg, "lat": LATS_reg, "depth":gridded_z, "time": unique_years}).T #fixes lat-lon bug
```

```python
for (tt, y) in enumerate(unique_years):
    print(tt / len(unique_years))
    which_year_and_notnan = (years == y) * (~np.isnan(cfc12))
    
    depths_ = depths[which_year_and_notnan]
    months_ = months[which_year_and_notnan]
    days_ = days[which_year_and_notnan]
    minutes_ = minutes[which_year_and_notnan]
    times_ = times[which_year_and_notnan]
    lat_  = lat[which_year_and_notnan]
    lon_ = lon[which_year_and_notnan]
    cfc12_ = cfc12[which_year_and_notnan]
    cast_codes_ = cast_codes[which_year_and_notnan]
    cast_num_ = cast_num[which_year_and_notnan]
    all_casts_ = all_casts[which_year_and_notnan]
    # cast_id = cast_codes + cast_num
    casts_ = list(set(all_casts_.values.tolist()))
    
    regular_cfc = np.zeros((len(gridded_z), len(casts_)))
    LATS = np.zeros(len(casts_))
    LONS = np.zeros(len(casts_))
    
    for j in range(len(casts_)):
        which_cast = (all_casts_ == casts_[j])
        unique_depths = np.unique(depths_[which_cast])
        time_averaged = np.zeros(len(unique_depths))
        LATS[j] = lat_[which_cast][0]
        LONS[j] = lon_[which_cast][0]
        for (i, d) in enumerate(unique_depths):
            time_averaged[i] = np.nanmean(cfc12_[which_cast][depths_[which_cast] == d])
        f = interpolate.interp1d(unique_depths, time_averaged, fill_value = np.nan, bounds_error = False)
        regular_cfc[:, j] = f(gridded_z)
    
    
    for k in range(len(gridded_z)):
        binning = pyinterp.Binning2D(
            pyinterp.Axis(LONS_reg, is_circle=True),
            pyinterp.Axis(LATS_reg))
        binning.clear()
        binning.push(LONS, LATS, 1 * regular_cfc[k, :][:], True)        
        cfcs_gridded_nearest[:, :, k, tt] = 1 * binning.variable('mean')
```

```python
cfc_gridded = cfc_gridded.compute()
```

```python
cfc_gridded.to_netcdf("GLODAP_gridded_CFC.nc", compute = True)
```
