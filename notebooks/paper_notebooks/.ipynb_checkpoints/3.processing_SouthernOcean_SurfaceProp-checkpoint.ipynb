### Load in data

```python
datadir = lambda x="" : "/vortexfs1/home/anthony.meza/scratch/CM4XTransientTracers/data/model/budgets_sigma2/" + x
datafiles = glob.glob(datadir("CM4Xp125*"))[20:]
datafiles = sorted(datafiles)

variables = ["tos", 
             "sos", 
             "siconc", 
             "LSNK", 
             "LSRC"]

dss = []

for (t, file) in enumerate(datafiles): 
    print(file)
    ds = xr.open_mfdataset(
        file,
        data_vars="minimal",
        coords="minimal",
        compat="override",
        parallel=True,
        engine="zarr")
    ds = ds[variables]
    ds['ant_mask'] = (ds['geolat'] <= -40)

        
    ds = ds.compute()
    dss += [1 * ds]
```

```python
wmts_ds = xr.concat(dss, dim = "time")
savedir = "/vortexfs1/home/anthony.meza/scratch/CM4XTransientTracers/data/model/"
wmts_ds.to_netcdf(savedir + "Southern_Ocean_Surface_Properties.nc")
```

```python
def area_average(ds, varname, mask = None):
    if mask is None:
        mask = 1
        
    numerator = mask * ds[varname] * ds["areacello"] * ds["wet"]
    denominator = mask * ds["areacello"] * ds["wet"]
    numerator = numerator.where(denominator > 0)
    denominator = denominator.where(denominator > 0)
    return numerator.sum(["xh", "yh"]) / denominator.sum(["xh", "yh"])
```

```python
(wmts_ds.tos * wmts_ds.ant_mask).mean(["xh", "yh"]).isel(exp = 0).groupby("time.year").mean("time").plot()
(wmts_ds.tos * wmts_ds.ant_mask).mean(["xh", "yh"]).isel(exp = 1).groupby("time.year").mean("time").plot()
```

```python
tos_average = area_average(wmts_ds, "tos", wmts_ds.ant_mask)
tos_average.isel(exp = 0).groupby("time.year").mean("time").plot()
tos_average.isel(exp = 1).groupby("time.year").mean("time").plot()
```

```python
tos_average = area_average(wmts_ds, "siconc", wmts_ds.ant_mask)
tos_average.isel(exp = 0).groupby("time.year").mean("time").plot()
tos_average.isel(exp = 1).groupby("time.year").mean("time").plot()
```
