### Load in data

```python
datadir = lambda x="" : "/vortexfs1/home/anthony.meza/scratch/CM4X/CM4XTransientTracers/data/model/budgets_sigma2/" + x
datafiles = glob.glob(datadir("CM4Xp125*"))[20:]
datafiles = sorted(datafiles)

wmts = []

for (t, file) in enumerate(datafiles[30:32]): 
    print(file)
    ds = xr.open_mfdataset(
        file,
        data_vars="minimal",
        coords="minimal",
        compat="override",
        parallel=True,
        engine="zarr")
    ds = ds.fillna(0.)
```

```python
ds.sel(sigma2_l = 37.22, method = "nearest").isel(time = 0, exp = 1).thkcello.plot(vmin = 0, vmax = 10)
```
